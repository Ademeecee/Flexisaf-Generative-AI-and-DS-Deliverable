```python
# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Line Plot and Scatter Plot using Matplotlib

```python
# Create dummy data
np.random.seed(42)
x = np.arange(0, 10, 0.5)
y = np.sin(x)
```

```python
# Line plot
plt.figure(figsize=(8, 4))
plt.plot(x, y, label='sin(x)', color='blue', marker='o')

# Customise the plot
plt.title('Line Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.4)

# Display the plot
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(8, 4))
plt.scatter(x, y, color='green', marker='x')

# Customise the plot
plt.title('Scatter Plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid(True, linestyle='--', alpha=0.4)

# Display the plot
plt.show()
```

#### Histogram and Box Plot using Seaborn

```python
# Create a sample dataset for Seaborn
df = pd.DataFrame({
    'Age': np.random.normal(30, 5, 100),
    'Salary': np.random.normal(50000, 8000, 100),
    'Department': np.random.choice(['HR', 'Tech', 'Marketing'], 100)
})

# Check first 5 rows
df.head()
```

```python
# Histogram
plt.figure(figsize=(8, 4))
sns.histplot(df['Age'], kde=True, color='purple')

# Customise the plot
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Count')

# Display the plot
plt.show()
```

**Observation**:
Most age are between 25 and 35.

```python
# Box Plot of Tips by Day and Time
plt.figure(figsize=(8, 4))
sns.boxplot(x='Department', y='Salary', data=df)

# Customise the plot
plt.title('Box Plot of Salary by Department')
plt.xlabel('Department')
plt.ylabel('Salary')

# Display the plot
plt.show()
```

**Observations**:

- **Median Salary**: HR has the highest median, followed closely by Marketing. Tech is lower.

- **Outliers**:

  - Marketing has a low-end outlier (~22,000).

  - Tech has a high-end outlier (~70,000).

- **Spread**: HR has a tight, consistent spread while Tech and Marketing have wider ranges.
